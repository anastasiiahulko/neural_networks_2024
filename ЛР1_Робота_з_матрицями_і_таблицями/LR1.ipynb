```python
import numpy as np

# a. Створити квадратну матрицю (n × n), заповнити її за правилом a_(ij)=(2j-0.5i)^(3)
n = 4
matrix_a = np.zeros((n, n))
for i in range(n):
    for j in range(n):
        matrix_a[i, j] = (2 * (j + 1) - 0.5 * (i + 1)) ** 3

print("Matrix A:")
print(matrix_a)

# Знайти детермінант матриці
det_a = np.linalg.det(matrix_a)
print("\nDeterminant of matrix A:", det_a)

# Знайти обернену матрицю
if det_a != 0:
    inverse_a = np.linalg.inv(matrix_a)
    print("\nInverse of matrix A:")
    print(inverse_a)
else:
    print("\nMatrix A is singular and does not have an inverse.")

# b. Задати два вектори з n елементами, заповнені випадковими числами
vector_real = np.random.rand(n)  # Вектор з дійсними числами
vector_int = np.random.randint(1, 10, size=n)  # Вектор з цілими числами

print("\nVector with real numbers:", vector_real)
print("Vector with integer numbers:", vector_int)

# Обчислити їх суму і добуток
vector_sum = vector_real + vector_int
vector_dot_product = np.dot(vector_real, vector_int)

print("\nSum of vectors:", vector_sum)
print("Dot product of vectors:", vector_dot_product)

# c. Задати дві матриці, заповнені випадковими числами: одна розміру n × m, друга – m × n
m = 3
matrix_b = np.random.rand(n, m)
matrix_c = np.random.rand(m, n)

print("\nMatrix B (n x m):")
print(matrix_b)
print("\nMatrix C (m x n):")
print(matrix_c)

# Знайти їх добуток
matrix_product_bc = np.dot(matrix_b, matrix_c)

print("\nProduct of matrix B and matrix C:")
print(matrix_product_bc)

# d. Знайти добуток матриць і векторів (розбити на дві пари, для яких можливо виконання операції)
# Матриця A і вектор реал
matrix_vector_product1 = np.dot(matrix_a, vector_real)
print("\nProduct of matrix A and vector_real:")
print(matrix_vector_product1)

# Матриця C і вектор цілих чисел
matrix_vector_product2 = np.dot(matrix_c, vector_int)
print("\nProduct of matrix C and vector_int:")
print(matrix_vector_product2)
```
